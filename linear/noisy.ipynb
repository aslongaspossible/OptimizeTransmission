```python
import matplotlib.pyplot as plt
```

```python
from plot import multi_plot
```

```python
def plot_noise(axes=None):
    if axes is None:
        axes = plt.subplots(2,2, figsize=(20, 17))
    fig, ((ax0, ax1), (ax2,ax3)) = axes
    data_list = [
        data0,
        data03,
        data1,
        data3,
    ]
    label_list = [
        0,
        .03,
        0.1,
        .3,
    ]
    line_style_list = [
        'k-',
        '--',
        '-.',
        ':',
    ]
    # x_range = np.linspace(-3, 3, 301)
    multi_plot(
        data1[:,0], 
        [x[:, 3] for x in data_list], 
        label_list, line_style_list, (fig, ax0)
    )
    ax0.set_ylabel(r'$zT$')
    ax0.set_ylim([0,5])
    ax0.set_yticks([1, 2, 3, 4, 5])
    multi_plot(
        data1[:,0], 
        [x[:, 5] for x in data_list], 
        label_list, line_style_list, (fig, ax1)
    )
    ax1.set_ylabel(r'$S{\cdot}e/k_\mathrm{B}$')
    # ax1.set_ylim([0, 8])
    # ax1.set_yticks([4, 8])
    multi_plot(
        data1[:,0], 
        [x[:, 6] for x in data_list], 
        label_list, line_style_list, (fig, ax2)
    )
    ax2.set_ylabel(r'$G{\cdot}h/e^2$')
    # ax2.set_ylim([0, 1])
    # ax2.set_yticks([.2, .4, .6, .8, 1.])
    ax2.set_xlabel(r'$1/\beta$')
    ax0.set_xlim([0,7])
    ax1.set_xlim([0,7])
    ax2.set_xlim([0,7])
    ax3.set_xticks([])
    ax3.set_yticks([])
    # ax0.legend('', frameon=False)
    ax1.legend('', frameon=False)
    ax2.legend('', frameon=False)
    # ax2.legend(fontsize=30)
_ = plot_noise()
```
